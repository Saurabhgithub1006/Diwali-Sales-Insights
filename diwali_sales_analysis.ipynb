# Exploratory Data Analysis

## Obsessed Diwali buyers  With respect to Gender

```python
ax = sns.countplot(x = "Gender",data =df)
```

```python
plt.figure(figsize =(10,10))

ax = sns.countplot(x = "Gender",data =df)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
 df["Gender"].value_counts()
```

## Expensive Shopping with respect to Gender

```python
Gen_sales = df.groupby(["Gender"] , as_index = False)["Amount"].sum().sort_values(["Amount"],ascending = False)
Gen_sales
```

```python
plt.figure(figsize =(10,10))

sns.barplot(x = "Gender" , y = "Amount" , data = Gen_sales )
```

##### From this we get that most of the buyers are females and also the purchasing amount of products is greater in females than males.

## Shopoholic Age Group

```python
df.columns
```

```python
plt.figure(figsize =(16,10))
ax =sns.countplot(x="Age Group",data = df,hue="Gender")

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
df["Age Group"].value_counts()
```

```python
df.groupby(["Age Group"] , as_index = False)["Amount"].sum().sort_values(["Amount"] ,ascending = False)
```

```python
Age_Group_sales =df.groupby(["Age Group"] , as_index = False)["Amount"].sum().sort_values(["Amount"] ,ascending = False)
```

```python
plt.figure(figsize =(16,10))


ax =sns.barplot(x = "Age Group" , y="Amount" ,data = Age_Group_sales)
 
ax.bar_label(ax.containers[0])    
plt.plot()
```

 ##### From this we see that the maximum buyers are from age group 26 - 35 and mostly females.

## Most obsessed Diwali shopping State

```python
df.head()
```

### ranking of states with orders

```python
state_orders = df.groupby(["State"],as_index =False)["Orders"].sum().sort_values(["Orders"],ascending =False)
state_orders
```

```python
plt.figure(figsize = (26,10))
ax =sns.barplot(x="State",y="Orders",data = state_orders)

ax.bar_label(ax.containers[0])
plt.plot()
```

### Ranking of States with Amount

```python
states_amount = df.groupby(["State"],as_index =False)["Amount"].sum().sort_values(["Amount"],ascending=False)
states_amount
```

```python
plt.figure(figsize =(26,10))
ax = sns.barplot(x="State",y ="Amount", data =states_amount )

ax.bar_label(ax.containers[0])
plt.plot()
```

#####  From this we get that most of the orders  and sales amount are from UP, Maharashtra, Karnataka 

## Passionate buyers profession

```python
df.head()
```

```python
occu_amount = df.groupby(["Occupation"],as_index = False)["Amount"].sum().sort_values(["Amount"],ascending = False)
```

```python
ax =sns.barplot(x= "Occupation", y ="Amount", data = occu_amount)
ax.bar_label(ax.containers[0])
plt.plot()
```

##### From the above we can see that the purchasing power with profession is high from IT Sector, Healthcare, Aviation and so on.
